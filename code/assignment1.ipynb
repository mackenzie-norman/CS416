```python

from scipy.io.wavfile import write
import sounddevice
import math
import matplotlib.pyplot as plt
import numpy as np
#arr = []
#high_val = 32767
#high_val /= 4
fs = 440
SAMPLE_RATE = 48000
t = np.linspace(0., 1., SAMPLE_RATE)
amplitude = (16384. )
data = amplitude * np.sin( 2 * math.pi * fs * t)
data = data.clip(min = (-1 * amplitude/2), max=amplitude/2 )
data = data.astype(np.int16)
write('clipped.wav', SAMPLE_RATE, data)
plt.plot(data[0: 4 * int(48000/440)])
plt.show()
sounddevice.play(data, SAMPLE_RATE)
```
